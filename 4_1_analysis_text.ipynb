# Import

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Load data

```python
df = pd.read_csv('assets/text_analysis.csv')

print(df.shape)
```

```python
df = pd.read_excel('filtered/text.xlsx')

print(df.shape)
```

# Counter

```python
df.groupby('participant_id').aggregate({'id': 'count'}).sort_values('id', ascending=False).head(10)
```

# Time

```python
ax = df.groupby('date').aggregate({'id': 'count'}).plot(marker='o')
plt.xlabel('Date')
plt.ylabel('Utilizzo')
plt.xticks(rotation=45, ha='right')
plt.legend([])
plt.show()
```
